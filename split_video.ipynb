## Splitting videos into frame in single folder code :

```python
#splitting multiple videos into frames in single folder
import os
import cv2
src = 'D:/internship/dataset/sample'
dest = 'D:/internship/dataset/split images'
for root, dirs, files in os.walk(src):
    frameNr = 0 #setting the frame counter to 0 outside the loop so it doesnt overwrite existing images
    gap = 10 #setting a gap variable to skip frames and print every 10 frames
    for file in files:
        if file.endswith('mp4'):
            capture = cv2.VideoCapture(f'D:/internship/dataset/sample/{file}')
            while (True):
                success, frame = capture.read()
                if success == False:
                    capture.release()
                    break

                if frameNr == 0:
                    cv2.imwrite(f'D:/internship/dataset/split images/frame_{frameNr}.jpg', frame)
                else:
                    if frameNr % gap == 0:
                        cv2.imwrite(f'D:/internship/dataset/split images/frame_{frameNr}.jpg', frame)
                frameNr = frameNr+1
            
```

## Splitting multiple videos into frame in different folders 

```python
import os
import numpy as np
import cv2
from glob import glob

def create_dir(path):
    try:
        if not os.path.exists(path):
            os.makedirs(path)
    except OSError:
        print(f"ERROR: creating directory with name {path}")

def save_frame(video_path, save_dir, gap=10):
    name = video_path.split("/")[-1].split(".")[0]
    save_path = os.path.join(save_dir, name)
    create_dir(save_path)

    cap = cv2.VideoCapture(video_path)
    idx = 0

    while True:
        ret, frame = cap.read()

        if ret == False:
            cap.release()
            break

        if idx == 0:
            cv2.imwrite(f"{save_path}/{idx}.png", frame)
        else:
            if idx % gap == 0:
                cv2.imwrite(f"{save_path}/{idx}.png", frame)

        idx += 1

if __name__ == "__main__":
    video_paths = glob("D:/internship/dataset/sample/*")
    save_dir = "save"

    for path in video_paths:
        save_frame(path, save_dir, gap=10)

```
